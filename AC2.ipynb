```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = '/content/pokemon_team_rocket_dataset.csv'
df = pd.read_csv(file_path)

selected_data = df.iloc[1:, [6, 9]].dropna()

plt.figure(figsize=(10, 7))
plt.boxplot([selected_data.iloc[:, 0], selected_data.iloc[:, 1]],
            labels=selected_data.columns)

plt.title('Boxplot')
plt.xlabel('Columnas')
plt.ylabel('Valores')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/pokemon_team_rocket_dataset.csv'
df = pd.read_csv(file_path)


selected_data = df.iloc[1:, [6, 9]]
sns.histplot(data=selected_data, x=selected_data.columns[0], kde=True)
plt.title('Histogram of ' + selected_data.columns[0])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
